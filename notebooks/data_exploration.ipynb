```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("akshatshaw7/moives-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd

movies = pd.read_csv(f"{path}/MOIVES_DATA.csv")
genres = pd.read_csv(f"{path}/genres.csv")
```

```python
movies.info()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
import ast
from typing import List

genre_dict = dict(zip(genres["id"], genres["name"]))

# Convert stringified lists to actual lists
movies['genre_ids'] = movies['genre_ids'].apply(ast.literal_eval)

def map_genre_ids_to_names(genre_ids: List) -> List[str]:
    return [genre_dict.get(int(gid), "Unknown") for gid in genre_ids]

movies['genre_names'] = movies['genre_ids'].apply(map_genre_ids_to_names)
```

```python
movies
```

```python
movies["tagged_overview"] = movies[["id", "overview"]].astype(str).agg(' '.join, axis=1)
movies
```

```python
movies.to_csv("../data/movies_cleaned_with_genres.csv", index=False)
```

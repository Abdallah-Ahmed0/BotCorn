```python
import pandas as pd

movies = pd.read_csv("data/movies_with_posters.csv")
```

```python
movies
```

```python
from itertools import chain

all_genres = set(chain.from_iterable(movies['genre_names']))

# Sort them for readability
sorted_genres = sorted(all_genres)

# Join into a clean comma-separated string
genres_str = ", ".join(sorted_genres)

# Print result
print(genres_str)
```
